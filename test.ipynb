```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv("sample_data/california_housing_test.csv")
```

```python
df.sample(3)
```

```python
age_1000_2000 = df[
    (df["housing_median_age"] > 10)
    &
     (df["housing_median_age"] < 20)
]
```

```python
age_1000_2000.groupby("housing_median_age").agg({
    "median_income":"mean",
    "median_house_value":"mean"
}).reset_index()
```

```python
age_1000_2000.to_parquet("test_age.parquet")
```
